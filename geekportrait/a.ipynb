```python
import numpy as np
```

```python
files = [x for x in open('a.txt').readlines()]
from PIL import Image
pics = []
for fn in files:
  pics.append(np.array(Image.open(fn[:-1])))


```

```python
p = np.array(pics)
p.shape
```

```python
av = np.average(p,axis=0)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(av.astype(np.int))
```

```python
Image.fromarray(av.astype(np.uint8)).save("geek_port.png")
```

```python
av = av/190*255
plt.imshow(av.astype(np.int))
```

```python
Image.fromarray(av.astype(np.uint8)).save("geek_port_n.png")
```

```python
import cv2
```

```python
video = cv2.VideoWriter('video2.avi', 0, 15, (300,300))
for i in range(564):
  p = np.array(pics[i:i+10])
  mx = np.max(p)
  av = np.average(p,axis=0)
  av = av/mx*255
  av = av.astype(np.uint8)
  av = cv2.cvtColor(av, cv2.COLOR_RGB2BGR)
  #Image.fromarray(av.astype(np.uint8)).save(f'/Users/vickievalerie/Desktop/2022-07-30/pictures/geek_port_{i}.png')
  video.write(av)

cv2.destroyAllWindows()
video.release()



  
```
